```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Đọc ảnh RGB
data = iio.imread('fruit.jpg')
print("Kích thước ảnh gốc:", data.shape)

# Tạo hàm zoom từng kênh
def zoom_rgb_image(image, zoom_factor):
    zoomed_channels = [nd.zoom(image[:, :, i], zoom_factor) for i in range(3)]
    return np.stack(zoomed_channels, axis=2)

# Phóng to gấp 3 lần
zoom_in = zoom_rgb_image(data, 3)
print("Ảnh phóng to (3x):", zoom_in.shape)

# Thu nhỏ 0.3 lần
zoom_out = zoom_rgb_image(data, 0.3)
print("Ảnh thu nhỏ (0.3x):", zoom_out.shape)

# Hiển thị so sánh
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(data)
plt.title("Ảnh gốc")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(zoom_in.astype(np.uint8))  # Chuyển về uint8 để hiển thị đúng
plt.title("Phóng to 3x")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(zoom_out.astype(np.uint8))
plt.title("Thu nhỏ 0.3x")
plt.axis('off')

plt.tight_layout()
plt.show()

# | Thuộc tính    | Phóng to (3x)                | Thu nhỏ (0.3x)                |
# | ------------- | ---------------------------- | ----------------------------- |
# | Kích thước    | Tăng mạnh                    | Giảm mạnh                     |
# | Độ chi tiết   | Mờ hơn, có thể bị rỗ nhẹ     | Mất chi tiết, bị nén          |
# | Hiển thị      | Rõ hơn nhưng ít sắc nét      | Nhỏ, khó quan sát             |
# | Mục đích dùng | Phân tích, quan sát chi tiết | Tạo preview, tiết kiệm bộ nhớ |
```
