```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# Đọc ảnh gốc
data = iio.imread('fruit.jpg')
print("Kích thước ảnh gốc:", data.shape)

# Phóng to gấp 3 lần
zoom_in = nd.zoom(data, 3)
print("Ảnh phóng to (gấp 3):", zoom_in.shape)

# Thu nhỏ 0.3 lần
zoom_out = nd.zoom(data, 0.3)
print("Ảnh thu nhỏ (0.3x):", zoom_out.shape)

# Hiển thị so sánh
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(data)
plt.title("Ảnh gốc")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(zoom_in)
plt.title("Phóng to 3x")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(zoom_out)
plt.title("Thu nhỏ 0.3x")
plt.axis('off')

plt.tight_layout()
plt.show()
```
