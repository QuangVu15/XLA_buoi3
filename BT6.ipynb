```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# c nh grayscale
data = iio.imread('world_cup.jpg', mode='F')  # Float grayscale
print(data.shape)
V, H = data.shape

# To ma trn ch s pixel
M = np.indices((V, H))

#  lch nhiu
d = 5
q = 2 * d * np.random.rand(*M.shape) - d  # to nhiu ngu nhin t -d n d

# To ch s mi vi nhiu
mp = (M + q).astype(int)

# Ly nh mi theo ch s  nhiu
dl = nd.map_coordinates(data, mp)

# Hin th nh
plt.imshow(dl, cmap='gray')
plt.show()

# nhn xt
# Bin i khng lm mt d liu m ch lm mo nh
```
