```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# Đọc ảnh grayscale
data = iio.imread('world_cup.jpg', mode='F')  # Float grayscale
print(data.shape)
V, H = data.shape

# Tạo ma trận chỉ số pixel
M = np.indices((V, H))

# Độ lệch nhiễu
d = 5
q = 2 * d * np.random.rand(*M.shape) - d  # tạo nhiễu ngẫu nhiên từ -d đến d

# Tạo chỉ số mới với nhiễu
mp = (M + q).astype(int)

# Lấy ảnh mới theo chỉ số đã nhiễu
dl = nd.map_coordinates(data, mp)

# Hiển thị ảnh
plt.imshow(dl, cmap='gray')
plt.show()

# nhận xét
# Biến đổi không làm mất dữ liệu mà chỉ làm méo ảnh
```
