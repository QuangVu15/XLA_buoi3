```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
import scipy.ndimage as nd

# Đọc ảnh gốc dưới dạng float
data = iio.imread('fruit.jpg', mode="F")

# Dịch ảnh theo hai hướng khác nhau
bdata1 = nd.shift(data, (100, 25))     # Dịch xuống 100, sang phải 25
bdata2 = nd.shift(data, (50, -30))     # Dịch xuống 50, sang trái 30

# Hiển thị ảnh gốc và hai ảnh sau dịch chuyển để so sánh
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(data, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(bdata1, cmap='gray')
plt.title('Shifted (100, 25)')

plt.subplot(1, 3, 3)
plt.imshow(bdata2, cmap='gray')
plt.title('Shifted (50, -30)')

plt.tight_layout()
plt.show()



# Nhận xét khi chạy với (50, -30):
# Phần ảnh gốc bị dời đi xuống dưới và sang trái, nên:

# Mép trên và phải của ảnh sẽ xuất hiện vùng màu đen hoặc xám đen (do thiếu dữ liệu, được nội suy bằng giá trị mặc định gần 0).

# Một phần ảnh bên trái và dưới có thể bị cắt mất.

# Chất lượng ảnh có thể bị mờ nhẹ do interpolation, vì nd.shift không chỉ dịch pixel nguyên mà còn nội suy giá trị giữa các pixel.
```
