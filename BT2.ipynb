```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
import scipy.ndimage as nd

# c nh gc di dng float
data = iio.imread('fruit.jpg', mode="F")

# Dch nh theo hai hng khc nhau
bdata1 = nd.shift(data, (100, 25))     # Dch xung 100, sang phi 25
bdata2 = nd.shift(data, (50, -30))     # Dch xung 50, sang tri 30

# Hin th nh gc v hai nh sau dch chuyn  so snh
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(data, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(bdata1, cmap='gray')
plt.title('Shifted (100, 25)')

plt.subplot(1, 3, 3)
plt.imshow(bdata2, cmap='gray')
plt.title('Shifted (50, -30)')

plt.tight_layout()
plt.show()



# Nhn xt khi chy vi (50, -30):
# Phn nh gc b di i xung di v sang tri, nn:

# Mp trn v phi ca nh s xut hin vng mu en hoc xm en (do thiu d liu, c ni suy bng gi tr mc nh gn 0).

# Mt phn nh bn tri v di c th b ct mt.

# Cht lng nh c th b m nh do interpolation, v nd.shift khng ch dch pixel nguyn m cn ni suy gi tr gia cc pixel.
```
