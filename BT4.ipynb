```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# Đọc ảnh gốc
data = iio.imread('fruit.jpg')
print("Kích thước ảnh gốc:", data.shape)

# Xoay 45 độ, reshape=True
d1 = nd.rotate(data, 45, reshape=True)
print("Kích thước ảnh reshape=True:", d1.shape)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(d1)
plt.title("Xoay 45° (reshape=True)")
plt.axis('off')

# Xoay 45 độ, reshape=False
d2 = nd.rotate(data, 45, reshape=False)
print("Kích thước ảnh reshape=False:", d2.shape)
plt.subplot(1, 2, 2)
plt.imshow(d2)
plt.title("Xoay 45° (reshape=False)")
plt.axis('off')

plt.tight_layout()
plt.show()

# reshape=True: giúp hiển thị ảnh đầy đủ sau xoay, tránh mất mát.
# reshape=False: giữ nguyên khung, nhưng có thể bị cắt góc ảnh.
```
