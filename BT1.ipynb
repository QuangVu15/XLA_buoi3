```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt

# Đọc ảnh gốc
data = iio.imread('fruit.jpg')
print("Kích thước ảnh:", data.shape)

# Cắt vùng ảnh quả cam và quả táo
orange = data[800:1200, 570:980]       # Quả cam
apple  = data[850:1150, 1550:1850]     # Quả táo

# Lưu ảnh
iio.imsave('orange.png', orange)
iio.imsave('apple.jpg', apple)

# Hiển thị cả hai ảnh
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(orange)
plt.title("Quả cam")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(apple)
plt.title("Quả táo")
plt.axis('off')

plt.tight_layout()
plt.show()
```
